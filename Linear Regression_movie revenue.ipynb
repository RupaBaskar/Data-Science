```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv(r"C:\Users\Malar\Rupa\cost_revenue_clean.csv")
```

```python
data
```

```python
data.describe()
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.title("Film_cost vs Gobal_Revenue")
plt.xlabel('production_budget_usd')
plt.ylabel('worldwide_gross_usd')
plt.xlim(0, 425000000.0)
plt.ylim(0, 2783919000.0)
plt.scatter(X,y, alpha=0.3)
plt.show()
```

```python
regression = LinearRegression()
```

```python
regression.fit(X, y)
```

```python
regression.coef_ #slope of the coefficient
```

```python
regression.intercept_ #Intercept
```

```python
plt.figure(figsize=(10,6))
plt.title("Film_cost vs Gobal_Revenue")
plt.xlabel('production_budget_usd')
plt.ylabel('worldwide_gross_usd')
plt.scatter(X,y, alpha=0.3)
plt.plot(X, regression.predict(X), color='Red', linewidth=5)
plt.xlim(0, 425000000.0)
plt.ylim(0, 2783919000.0)
plt.show()
```

```python
regression.score(X, y) #Getting r square from Regression
```
