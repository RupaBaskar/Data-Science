```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
from scipy import stats
```

```python
a = sns.load_dataset("flights") #you can directly get the dataset from github
sns.relplot(x="passengers", y="month", data=a) #relplot() use to visualize statistical relationship between varaibale in dataset
```

```python
a = sns.load_dataset("flights")
sns.relplot(x="passengers", y="month",hue="year",data=a)
```

Plotting with categorical data

```python
b = sns.load_dataset("tips")
sns.relplot(x="time", y="tip", data=b, kind="line")
```

```python
sns.catplot(x="day", y="total_bill", data=b)
```

```python
sns.catplot(x="day", y="total_bill", data=b, kind="violin")
```

```python
sns.catplot(x="day", y="total_bill", data=b, kind="boxen")
```

```python
c=np.random.normal(loc=5,size=100,scale=2)
sns.distplot(c)
```

Multiplot Grids

```python
a=sns.load_dataset("iris")
b=sns.FacetGrid(a, col="species")
b.map(plt.hist, "sepal_length")
```

```python
b=sns.load_dataset("flights")
c=sns.PairGrid(b)
c.map(plt.scatter)
```

Plot Aesthetics

```python
sns.set(style="darkgrid")
a=sns.load_dataset("iris")
b=sns.FacetGrid(a, col="species")
b.map(plt.hist, "sepal_length")
```

```python
sns.set(style="dark")
a=sns.load_dataset("iris")
b=sns.FacetGrid(a, col="species")
b.map(plt.hist, "sepal_length")
```

```python
sns.set(style="white", color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=a)
```

```python
sns.set(style="white", color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=a)
sns.despine(offset=10, trim=True)
```
